```python
# 19.4 Linear Regression Dataset
from numpy import array
from matplotlib import pyplot

# define dataset
dataset = array([
    [0.05,0.12],
    [0.18,0.22],
    [0.31,0.35],
    [0.42,0.38],
    [0.5,0.49]
])

print(dataset)

# split into inputs and outputs
X, y = dataset[:,0], dataset[:,1]
X = X.reshape((len(X), 1))

# scatter plot
pyplot.scatter(X,y)
pyplot.show()
```

```python
# direct solution to linear least squares
from numpy import array
from numpy.linalg import inv
from matplotlib import pyplot

# define dataset
dataset = array([
    [0.05,0.12],
    [0.18,0.22],
    [0.31,0.35],
    [0.42,0.38],
    [0.5,0.49]
])

# split into inputs and outputs
X, y = dataset[:,0], dataset[:,1]
X = X.reshape((len(X), 1))

# linear least squares
b = inv(X.T.dot(X)).dot(X.T).dot(y)
print(b)

# predict using coefficients
yhat = X.dot(b)

# plot data and predictions
pyplot.scatter(X,y)
pyplot.plot(X, yhat, color='red')
pyplot.show
```

```python
# QR decomposition solution to linear least squares
'''The QR decomposition approach is more computationally efficient and more numerically
stable than calculating the normal equation directly, but does not work for all data matrices.'''
from numpy import array
from numpy.linalg import inv
from numpy.linalg import qr
from matplotlib import pyplot

# define dataset
dataset = array([
    [0.05,0.12],
    [0.18,0.22],
    [0.31,0.35],
    [0.42,0.38],
    [0.5,0.49]
])

# split into input and output
X, y = dataset[:,0], dataset[:,1]
X = X.reshape((len(X),1))

# factorize
Q, R = qr(X)
b = inv(R).dot(Q.T).dot(y)
print(b)

# predict using coefficients
yhat = X.dot(b)

# plot data and predictions
pyplot.scatter(X,y)
pyplot.plot(X, yhat, color='red')
pyplot.show()
```

```python
# SVD solution via psuedoinverse to linear least squares

'''The pseudoinverse via SVD approach to solving linear least squares is the de facto standard
'''
from numpy import array
from numpy.linalg import pinv
from matplotlib import pyplot

# define dataset
dataset = array([
    [0.05,0.12],
    [0.18,0.22],
    [0.31,0.35],
    [0.42,0.38],
    [0.5,0.49]
])

# split into inputs and outputs
X, y = dataset[:, 0], dataset[:, 1]
X = X.reshape((len(X),1))

# calculate coefficients
b = pinv(X).dot(y)
print(b)

# predict using coefficients
yhat = X.dot(b)

# plot data and predictions
pyplot.scatter(X, y)
pyplot.plot(X, yhat, color='red')
pyplot.show()
```

```python
# least squares via convenience function
from numpy import array
from numpy.linalg import lstsq
from matplotlib import pyplot

# define dataset
dataset = array([
    [0.05,0.12],
    [0.18,0.22],
    [0.31,0.35],
    [0.42,0.38],
    [0.5,0.49]
])

# split into inputs and outputs
X, y = dataset[:,0], dataset[:,1]
X = X.reshape((len(X), 1))
print(len(X))
# calculate coefficients
b, residuals, rank, s = lstsq(X, y, rcond=-1)
print(b)

# predict using coefficients
yhat = X.dot(b)

# plot data and predictions
pyplot.scatter(X, y)
pyplot.plot(X, yhat, color='red')
pyplot.show()
```
